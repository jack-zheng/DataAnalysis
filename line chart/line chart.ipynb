```python
import csv
import plotly.plotly as py
import plotly.graph_objs as go

def get_data(file_path):
    content = []
    with open(file_path, 'r', newline='') as file:
        reader = csv.reader(file)
        content = list(reader)[1:]
    return content

data = get_data('data.csv')

count_cost_pair = {}
cost_index = 3
count_index = 2

for line in data:
    key = line[count_index]
    if not count_cost_pair.get(key):
        value  = [line[cost_index]]
        count_cost_pair[key] = value
    else: 
        count_cost_pair[key].append(line[cost_index])

# get valid points 
valid_x = []
valid_y = []
for sub in count_cost_pair:
    for t in count_cost_pair[sub]:
        valid_x.append((sub.split('-')[1]).replace('k', ''))
        valid_y.append(t)
        
valid_points = go.Scatter(
    x = valid_x,
    y = valid_y,
    mode = 'markers',
    name = 'valid',
    marker = dict(color = '#00CC66')
)

# get average points
avg_x = []
avg_y = []
for sub in count_cost_pair:
    avg_x.append((sub.split('-')[1]).replace('k', ''))
    
    tmp = 0.0
    for t in count_cost_pair[sub]:
        tmp = tmp + float(t)
        
    avg_time = round(tmp/len(count_cost_pair[sub]), 2)
    avg_y.append(avg_time)

avg_points = go.Scatter(
    x = avg_x,
    y = avg_y,
    mode = 'lines+markers',
    name = 'avg'
)


droped = go.Scatter(
    x = [800, 1000],
    y = [28.3, 39.75],
    mode = 'markers',
    name = 'droped',
    marker = dict(color = '#FF0000')
)
data = [valid_points, avg_points, droped]

layout = go.Layout(
    title='Hana Bulk Import Time Consuming',
    xaxis=dict(
        title='User Count/Thousand',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Time Cost/Mins',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Hana Bulk Import')
```
